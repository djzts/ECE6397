```
import os
from matplotlib import pyplot as plt
import cv2
import pydicom
import numpy as np
from tqdm import notebook
```

```
def turn_covid_to_npy(split=''):
    #root_for_covid = 'F:\\UoH\\ML\\data\\Covid\\'+split+'\\'   #F:\UoH\ML\data\Covid
    
    root_for_covid = '/content/drive/MyDrive/UHCourse/ECE6397/data/Covid/'
    
    #print(root_for_covid)
    size_list = [32, 56, 64, 112, 128, 256] #@param
    for size in size_list:
      for path in notebook.tqdm(os.listdir(root_for_covid), desc='ProgressCovid'+'_size_%d'%(size)):
          filename = path.split('.')[0]
          if path.split('.')[-1] == 'dcm':
              ds = pydicom.read_file(root_for_covid + path)
              size_2d= (size,size) 
              pix = np.stack((cv2.resize(ds.pixel_array, size_2d),) * 3, axis=-1)   
              pix = 2*(pix / 255) - 1
              #scale 1,-1
              try:
                os.makedirs('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Covid/' % size)
              except:
                pass
              np.save('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Covid/'%size+ filename +'.npy',pix)   #F:\UoH\ML\Data_npy\Covid
```

```
def turn_other_to_npy(split=''):
    #root_for_PNEUMONIA = 'D:/dataset/Other/'+split+'/PNEUMONIA'
    #root_for_NORMAL = 'D:/dataset/Other/'+split+'/NORMAL'
    #root_for_VIRUS = 'D:/dataset/Other/' + split + '/VIRUS'
    
    root_for_PNEUMONIA = '/content/drive/MyDrive/UHCourse/ECE6397/data/Bacteria/'
    root_for_NORMAL = '/content/drive/MyDrive/UHCourse/ECE6397/data/Normal/'
    root_for_VIRUS = '/content/drive/MyDrive/UHCourse/ECE6397/data/virus/'
    
    size_list = [32, 56, 64, 112, 128, 256] #@param
    for size in size_list:
      i=0
      for eachpath in [root_for_PNEUMONIA,root_for_NORMAL,root_for_VIRUS]:
          i=i+1
          for path in notebook.tqdm(os.listdir(eachpath), desc='Progress'+eachpath.split('/')[-2]+'_size_%d'%(size)):
              filename = path.split('.')[0] 
              if path.split('.')[-1] == 'jpeg':
                  jpg = cv2.imread(eachpath+path, 0)
                  size_2d = (size,size) 
                  pix = np.stack((cv2.resize(jpg, size_2d),) * 3, axis=-1)
                  pix = 2*(pix / 255) - 1
                           
                  if i==1:
                      try:
                        os.makedirs('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Bacteria/' % size)
                      except:
                        pass
                      np.save('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Bacteria/' % size+filename+'.npy',pix)
                
                  elif i==2:
                      try:
                        os.makedirs('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Normal/'% size)
                      except:
                        pass
                      np.save('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Normal/'% size + filename + '.npy', pix)
                
                  else:
                      try:
                        os.makedirs('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Virus/'% size)
                      except:
                        pass
                      np.save('/content/drive/MyDrive/UHCourse/ECE6397/Data_npy_%d/Virus/'% size + filename + '.npy', pix)
```
